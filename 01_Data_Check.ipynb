# Moving average

```python
total_mov_ave = []
for i in range(len(data)):
    temp = data.loc[i][1:]
    mov_ave = []
    for ii in range(1,len(data.columns)):
        mov_ave.append(np.average(temp[ii-1:ii+2]))
    y = mov_ave
    x = np.arange(len(y))
    total_mov_ave.append(y)
    plt.plot(x,y)
plt.show()
```
